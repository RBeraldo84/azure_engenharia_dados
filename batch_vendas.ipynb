## Carregando arquivo

```
vendas = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/contato@rodolfomoreira.com.br/vendas_finais.csv")
vendas.display()
```

## JDBC

```
driver = "org.mariadb.jdbc.Driver"

database_host = "vendas-petshop.mysql.database.azure.com"
database_port = "3306"
database_name = "vendas"
user = dbutils.secrets.get(scope="senhas-data-lake-house", key="vendas-user")
password = dbutils.secrets.get(scope="senhas-data-lake-house", key="vendas-password")
```

```
url = f"jdbc:mysql://{database_host}:{database_port}/{database_name}"
url
```

```
vendas.write \
    .format("jdbc") \
    .option("driver", driver) \
    .option("url", url) \
    .option("dbtable", "vendas_diarias") \
    .option("user", user) \
    .option("password", password) \
    .mode("overwrite").save()
```
