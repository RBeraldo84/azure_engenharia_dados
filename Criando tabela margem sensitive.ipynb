```
sql_margem = """SELECT 
  data_extracao,
  cod_loja,
  cod_produto,
  ROUND((margem / venda_quantidade), 2) as margem_produto
FROM
  work.posicao_estoque_work
WHERE
  margem > 0
LIMIT 100"""

df_margem = spark.sql(sql_margem)

df_margem.display()
```

```
df_margem.write \
    .option('path', '/mnt/sensitive/margem_produto_dia_sensitive2') \
    .partitionBy(['data_extracao', 'cod_loja', 'cod_produto']) \
    .mode('overwrite') \
    .saveAsTable('sensitive.margem_produto_dia_sensitive2')
```
